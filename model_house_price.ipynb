```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# drop the unwanted featuers from dataset for train the model
df = df.drop(columns = ['date', 'street', 'yr_renovated', 'city', 'statezip', 'country'])
df.head()
```

```python
print("hi")
```

```python
# find the outliers unwanted in the train model (noise of the model)
z_scores = stats.zscore(df)
threshold = 3
print("Size before removing outliers : ", df.shape)
outliers_df = df[(z_scores > threshold).any(axis=1)]
df = df[(z_scores <= threshold).all(axis=1)]
print("Size after removing outliers : ", df.shape)
```

```python
# print : removed first outliers values
outliers_df.head()
```

```python
# data preprocessing -> all values in same scales
scaler = StandardScaler()

scaler.fit(df)

df_scaled = pd.DataFrame(scaler.transform(df), columns=df.columns)
```

```python
# split the dataset
X = df.drop('price', axis=1)
y = df['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=12)
```

```python
print(f"X_train : {X_train.shape}")
print(f"X_test : {X_test.shape}")
```

```python
# create the model
models = [
    ('Random Forest : ', RandomForestRegressor()),
    ('Linear Regression : ', LinearRegression()),
    ('Decision Tree : ', DecisionTreeRegressor()),
    ('KNN : ', KNeighborsRegressor())
]
```

```python
for name, model in models:
    print(name)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print("Mean Squared Error : ", mean_squared_error(y_test, y_pred)) # its low good modle
    print("Mean Absolute Error : ", mean_absolute_error(y_test, y_pred)) # its low good modle
    print("R2 Squared : ", r2_score(y_test, y_pred)) # its high good modle 
    print()
```
